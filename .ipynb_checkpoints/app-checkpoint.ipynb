```python
#doing all necessary imports
from newspaper import Article
import networkx as nx
import matplotlib.pyplot as plt
import spacy
from spacy.lang.en import English
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from py2neo import Graph
```

```python
#url for scraping
url = 'https://thehimalayantimes.com/lifestyle/environment/hindu-kush-himalaya-region-under-threat-warn-experts/'
```

```python
#scraping article from above url
article = Article(url)
```

```python
#article download and parse
article.download()
article.parse()
```

```python
#getting string from the article
final_article = article.text
```

```python
#loading model from spacy
nlp = spacy.load("en_core_web_sm")
```

```python
#getting article paragragh, cleaning it, lemmatizing and returning list of sentences
def getSentences(text):
    nlp = English()
    nlp.add_pipe(nlp.create_pipe('sentencizer'))
    document = nlp(text)
    sentence_list = []
    for sentence in document.sents:
        sent = []
        senten = ''
        for word in sentence:
            if not word.is_punct:
                if '\n' not in word.text:
                    sent.append(word.lemma_)  
        senten = ' '.join(word for word in sent)      
        sentence_list.append(senten)
    return sentence_list
```

```python
#storing list of sentences by calling function in a variable
sentences = getSentences(final_article)
```

```python
#ist that contains subject and object dep_
SUBJECTS = ["nsubj", "nsubjpass", "csubj", "csubjpass", "agent", "expl"]
OBJECTS = ["dobj", "dative","pobj" "attr", "oprd"]
```

```python
#get subjects if there is 'and' to connect them
def getSubsFromConjunctions(subs):
    moreSubs = []
    for sub in subs:
        rights = list(sub.rights)
        rightDeps = {tok.lower_ for tok in rights}
        if "and" in rightDeps:
            moreSubs.extend([tok for tok in rights if tok.dep_ in SUBJECTS or tok.pos_ == "NOUN"])
            if len(moreSubs) > 0:
                moreSubs.extend(getSubsFromConjunctions(moreSubs))
    return moreSubs
```

```python
#get objects if there is 'and' to connect them
def getObjsFromConjunctions(objs):
    moreObjs = []
    for obj in objs:
        # rights is a generator
        rights = list(obj.rights)
        rightDeps = {tok.lower_ for tok in rights}
        if "and" in rightDeps:
            moreObjs.extend([tok for tok in rights if tok.dep_ in OBJECTS or tok.pos_ == "NOUN"])
            if len(moreObjs) > 0:
                moreObjs.extend(getObjsFromConjunctions(moreObjs))
    return moreObjs
```

```python
#get verbs if there is 'and' to connect them
def getVerbsFromConjunctions(verbs):
    moreVerbs = []
    for verb in verbs:
        rightDeps = {tok.lower_ for tok in verb.rights}
        if "and" in rightDeps:
            moreVerbs.extend([tok for tok in verb.rights if tok.pos_ == "VERB"])
            if len(moreVerbs) > 0:
                moreVerbs.extend(getVerbsFromConjunctions(moreVerbs))
    return moreVerbs
```

```python
#getting each subjects based on the verb and it's dependencies
def findSubs(tok):
    head = tok.head
    while head.pos_ != "VERB" and head.pos_ != "NOUN" and head.head != head:
        head = head.head
    if head.pos_ == "VERB":
        subs = [tok for tok in head.lefts if tok.dep_ == "SUB"]
        if len(subs) > 0:
            verbNegated = isNegated(head)
            subs.extend(getSubsFromConjunctions(subs))
            return subs, verbNegated
        elif head.head != head:
            return findSubs(head)
    elif head.pos_ == "NOUN":
        return [head], isNegated(tok)
    return [], False
```

```python
#processing negative tokens
def isNegated(tok):
    negations = {"no", "not", "n't", "never", "none"}
    for dep in list(tok.lefts) + list(tok.rights):
        if dep.lower_ in negations:
            return True
    return False
```

```python
#getting subject and verbs
def findSVs(tokens):
    svs = []
    verbs = [tok for tok in tokens if tok.pos_ == "VERB"]
    for v in verbs:
        subs, verbNegated = getAllSubs(v)
        if len(subs) > 0:
            for sub in subs:
                svs.append((sub.orth_, "!" + v.orth_ if verbNegated else v.orth_))
    return svs
```

```python
#getting objects if preposition is dependent
def getObjsFromPrepositions(deps):
    objs = []
    for dep in deps:
        if dep.pos_ == "ADP" and dep.dep_ == "prep":
            objs.extend([tok for tok in dep.rights if tok.dep_  in OBJECTS or (tok.pos_ == "PRON" and tok.lower_ == "me")])
    return objs
```

```python
#getting onjects from attributes
def getObjsFromAttrs(deps):
    for dep in deps:
        if dep.pos_ == "NOUN" and dep.dep_ == "attr":
            verbs = [tok for tok in dep.rights if tok.pos_ == "VERB"]
            if len(verbs) > 0:
                for v in verbs:
                    rights = list(v.rights)
                    objs = [tok for tok in rights if tok.dep_ in OBJECTS]
                    objs.extend(getObjsFromPrepositions(rights))
                    if len(objs) > 0:
                        return v, objs
    return None, None
```

```python
#get object from xcomp
def getObjFromXComp(deps):
    for dep in deps:
        if dep.pos_ == "VERB" and dep.dep_ == "xcomp":
            v = dep
            rights = list(v.rights)
            objs = [tok for tok in rights if tok.dep_ in OBJECTS]
            objs.extend(getObjsFromPrepositions(rights))
            if len(objs) > 0:
                return v, objs
    return None, None
```

```python
#get all the list of subjects and verb
def getAllSubs(v):
    verbNegated = isNegated(v)
    subs = [tok for tok in v.lefts if tok.dep_ in SUBJECTS and tok.pos_ != "DET"]
    if len(subs) > 0:
        subs.extend(getSubsFromConjunctions(subs))
    else:
        foundSubs, verbNegated = findSubs(v)
        subs.extend(foundSubs)
    return subs, verbNegated
```

```python
#get all the list of the objects and verb
def getAllObjs(v):
    rights = list(v.rights)
    objs = [tok for tok in rights if tok.dep_ in OBJECTS]
    objs.extend(getObjsFromPrepositions(rights))
    potentialNewVerb, potentialNewObjs = getObjFromXComp(rights)
    if potentialNewVerb is not None and potentialNewObjs is not None and len(potentialNewObjs) > 0:
        objs.extend(potentialNewObjs)
        v = potentialNewVerb
    if len(objs) > 0:
        objs.extend(getObjsFromConjunctions(objs))
    return v, objs
```

```python
#get subject, verb, object triple
def findSVOs(tokens):
    svos = []
    verbs = [tok for tok in tokens if tok.pos_ == "VERB" and tok.dep_ != "aux"]
    for v in verbs:
        subs, verbNegated = getAllSubs(v)
        if len(subs) > 0:
            v, objs = getAllObjs(v)
            for sub in subs:
                for obj in objs:
                    objNegated = isNegated(obj)
                    svos.append((sub.lower_, "!" + v.lower_ if verbNegated or objNegated else v.lower_, obj.lower_))
    return svos
```

```python
#print graph using networkx and store graph in database using neo4j
def printGraph(triple):
    G = nx.DiGraph()
    G.add_node(triple[0])
    G.add_node(triple[2])
    G.add_edges_from([(triple[0], triple[2])],title=triple[1])
    pos = nx.spring_layout(G)
    plt.figure   
    nx.draw(G, pos=pos, node_size=400, with_labels=True, fontsize=9,arrows=True)
    edge_labels = nx.get_edge_attributes(G, 'title')
    nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_labels)
    plt.axis('off')
    plt.show()
    g = Graph(password='1234')
    my_node = g.run("CREATE (a:Subject{name:$x})-[r:RELATES{name:$y}]->(b:Object{name:$z})",x=triple[0],y=triple[1],z=triple[2])

```

```python
#gives the triples for the sentences in article
def entitiesRelation(sentences):
    triple_list = []
    for sentence in sentences:
        sent = nlp(sentence)
        final_sent = findSVOs(sent)
        if len(final_sent)>0:
            for trip_list in final_sent:
                triple_list.append(list(trip_list))
                printGraph(list(trip_list))
    print(triple_list)   

    
```

```python
#calling function
entitiesRelation(sentences)
```

```python
#asking question
g = Graph(password='1234')
question_subject = ['attr',"nsubj", "nsubjpass", "csubj", "csubjpass"]
question_root = ['ROOT']
question_object = ['dobj',"dobj", "dative","pobj"]
ask = input('Ask me')
```

```python
#returns answer
a = nlp(ask)
for sent in a.sents:
    for word in sent:
        if word.dep_ in question_subject:
            subject = word.text
        elif word.dep_ in question_root:
            relation = word.text
        elif word.dep_ in question_object:
            obj = word.text
            
my_node = g.evaluate("MATCH (a)-[r]->(b) WHERE((a.name= $x AND b.name=$z) OR (r.name=$y AND b.name=$z) OR (a.name=$x AND r.name=$y)) RETURN a.name+' '+r.name+' '+b.name",x=subject,y=relation,z=obj)
print(my_node)
```
